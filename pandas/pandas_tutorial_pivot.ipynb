```python
import FinanceDataReader as fdr
```

```python
fdr.StockListing('KOSPI').head()
```

```python
samsung_price = fdr.DataReader('005930', '2024')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(samsung_price)
```

```python
df = df.reset_index()
```

```python
df['month'] = df.Date.dt.month
```

```python
df_summary = df[["month", "Close", "Volume"]].groupby('month').mean()
```

```python
df_summary= df_summary.reset_index()
```

```python
melted = df_summary.melt(id_vars=['month'], var_name = 'type', value_name='value')
```

```python
melted
```

```python
melted.pivot_table(index='month', values='value', columns='type')
```
