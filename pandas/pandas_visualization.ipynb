#### 이 예제는 같은 분산/평균/분위수를 가진 데이터가 얼마나 다를 수 있는지를 보여주는 대표적인 사례이다.

dataset 그룹별로 그룹바이를 해서 살펴보자

```python
data.groupby('dataset').describe()
```

```python

subset_I = data[data.dataset == 'I']
```

```python
# plt.plot 은 기본적으로 line graph 임 (들어온 순서대로 (x,y) 쌍을 선분으로 이음)
plt.plot( subset_I['x'], subset_I['y']) 
```

```python
## 점으로 찍고 싶으면
plt.plot( subset_I['x'], subset_I['y'], 'o') ## point 옵션을 주거나
plt.show()
#or 
plt.scatter( subset_I['x'], subset_I['y']) ## scatter chart 로 그리면 됨 
```

```python
## 불러온 데이터의 dataset 컬럼은 총 4개의 값으로 구분됨  다음을 통해 확인가능
data.dataset.unique()
```

```python
# 이를 각각 2 x 2 에 그려보기 연습 
# 먼저 격자를 만들고 (subplot 2, 2)
# 위부터 차례대로 1, 2 
#                3, 4 
# 번호를 받게 됨

fig = plt.figure() # 캔버스를 준비
axes1 = fig.add_subplot(2, 2, 1) ## 2x2 중에 1번째
axes2 = fig.add_subplot(2, 2, 2) ## 2x2 중에 2번째..
axes3 = fig.add_subplot(2, 2, 3) 
axes4 = fig.add_subplot(2, 2, 4)

```

이제 여기에 그림을 채우면 됨

```python
subset_I = data[data.dataset == 'I']
subset_II = data[data.dataset == 'II']
subset_III = data[data.dataset == 'III']
subset_IV = data[data.dataset == 'IV']
```

```python
axes1.plot( subset_I.x, subset_I.y, 'o')
axes2.plot( subset_II.x, subset_II.y, 'o')
axes3.plot( subset_III.x, subset_III.y, 'o')
axes4.plot( subset_IV.x, subset_IV.y, 'o')
fig
```

```python
## 꾸미기
```

```python
axes1.set_title("datset I")
axes2.set_title("datset II")
axes3.set_title("datset III")
axes4.set_title("datset IV")

fig
```
