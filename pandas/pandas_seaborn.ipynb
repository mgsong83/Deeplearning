```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head(5)
```

```python
sns.histplot( tips["tip"], kde=True) ## kde -> 밀도함수를 추정
```

```python
sns.kdeplot(tips["tip"])
```

```python
#날짜별로 팁 받은 날의 도수 분포 
sns.countplot( x = 'day', data=tips )
```

```python
sns.regplot( data = tips, x='total_bill', y='tip')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', hue='sex', data=tips)
```

```python
sns.kdeplot( data = tips, x = 'total_bill', y='tip', shade=True)
```

```python
sns.boxplot( data = tips, x='time', y='total_bill')
```

```python
# violinplot is better than box plot

sns.violinplot( data = tips, x='time', y='total_bill')
```

```python
sns.violinplot( data = tips, x='time', y='total_bill', hue='sex' , split=True)
```

```python
tips.head(5)
```

```python
sns.scatterplot( data=tips, x='total_bill', y='tip', hue='sex', style='time')
```

```python
g = sns.FacetGrid(tips, col='day', col_wrap=2)
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip', hue='sex')
```

```python
sns.lmplot( data=tips, x='total_bill', y='tip', hue='sex', col='day', col_wrap=2)
```

```python

sns.set_style('whitegrid')
g = sns.FacetGrid(tips, col='day', col_wrap=2)
g.map_dataframe(sns.scatterplot, x='total_bill', y='tip', hue='sex')
```
